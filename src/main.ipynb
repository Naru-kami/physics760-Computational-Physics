```python
import numpy as np
import matplotlib.pyplot as plt

def output(filename):
    return f"../figs/{filename}"
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x = np.sin(t)
y = np.cos(t)

plt.figure(figsize=(5,5))
# plt.tight_layout()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.savefig(output("test.png"), bbox_inches='tight')
```
